```python
#1
import pandas as pd
import seaborn as sns
data = sns.load_dataset('titanic')
data.head(10)
```

```python
#2
data.isna().sum()
```

```python
#4
average_age = data['age'].mean()
median_age = data['age'].median()
std_age = data['age'].std()

print(f"Average age: {average_age}")
print(f"Median age: {median_age}")
print(f"Standard deviation of age: {std_age}")
```

```python
#3
class_percentages = data['survived'].value_counts(normalize=True) * 100
print(class_percentages)
```

```python
#3
survived_by_gender = data.groupby('sex')['survived'].value_counts(normalize=True).mul(100).unstack()
print(survived_by_gender)
```

```python
#5
pclass_counts = data['pclass'].value_counts()
pclass_counts
```

```python
#6
female = data[(data['sex'] == 'female') & (data['age'] < 18) & (data['survived'] == 1)]
display(female)
```

```python
#7

median_age = data['age'].median()
data['age'].fillna(median_age, inplace=True)
display(data.head())
```

```python
#d
import numpy as np
data['age_group'] = np.where(data['age'] < 18, 'Child',
                             np.where(data['age'] <= 60, 'Adult', 'Senior'))
display(data.head())
```

```python
#9
top_5 = data.sort_values(by='fare', ascending=False).head(5)
display(top_5)
```
